```python
# Aplikacja do generowania cech technologicznych - typu rowki wpustowe
import numpy as np
import pandas as pd
# Generowanie liczb losowych
#from numpy.random import randn
#np.random.seed(101)
```

# Dane opisujące cechę technologiczną typu rowek wpustowy
D - Diameter: średnica powierzchni cylindrycznej (stopień wału) na której znajduje się rowek wpustowy; od 6 do 500
R - Radius: szerokość rowka = średnica freza: odpowiada parametrowi b; od 2 do 40
L - Length: długość rowka; od 6 do 500
G - Depth: głębokość rowka; od 1.2 do 31
Fit - pasowanie, dopuszczalne wartości: H9, N9, P9
IT - klasa dokładności, pole tolerancji: 9 lub 10
Ra - chropowatość powierzchni: od 0.63 do 10 (6.3)
Mat_PN - nazwa materiału obrabianego wg polskiej normy PN: 18 różnych
Rm - wytrzymałość materiału na rozciąganie w MPa: od 380 do 120 MPa
HB - twardość materiału wg Brinella: od 143 do 255

```python
# Wczytanie potrzebnych danych
material = pd.read_excel('qRowkiWpustowe.xlsx',sheet_name='material')
dfRa = pd.read_excel('qRowkiWpustowe.xlsx',sheet_name='Ra')
dfDims = pd.read_excel('qRowkiWpustowe.xlsx',sheet_name='wymiary')
dfLen = pd.read_excel('qRowkiWpustowe.xlsx',sheet_name='dlugosc')
```

```python
material.head()
```

```python
# Losowanie materiału
# Odczytać ile ich jest - poprzez zakres Nr
Mmin = material['Nr'].min()
Mmax = material['Nr'].max()
print(Mmin, Mmax)
MID = np.random.randint(Mmin,Mmax)
print(MID)
# Odczytaj materiał
Mat_Selected = material[material['Nr']==MID]
Mat_Selected
```

```python
Mat_PN = Mat_Selected['Mat_PN'].iloc[0]
print("Materiał PN: {0}".format(Mat_PN))
Rm = Mat_Selected['Rm'].iloc[0]
print("Wytrzymałość Rm = {0}".format(Rm))
HB = Mat_Selected['HB'].iloc[0]
print("Twardość HB = {0}".format(HB))
```

```python
# Zakres możliwych chropowatości
dfRa.head()
```

```python
# Losowanie chropowatości
Ramin = dfRa['Nr'].min()
Ramax = dfRa['Nr'].max()
print(Ramin,Ramax)
RaID = np.random.randint(Ramin,Ramax)
print(RaID)
# Odczytaj Ra
Ra_Selected = dfRa[dfRa['Nr']==RaID]
print(Ra_Selected)
Ra = Ra_Selected['Ra'].iloc[0]
print("Chropowatość Ra = {0}".format(Ra))
Fit = 'N9'
print("Pasowanie: {0}".format(Fit))
```

```python
# Zakresy możliwych wymiarów
dfDims.head()
```

```python
# Losowanie zestawu danych
DimMin = dfDims['Nr'].min()
DimMax = dfDims['Nr'].max()
print(DimMin, DimMax)
DID = np.random.randint(DimMin, DimMax)
print(DID)
# Odczytaj zestaw
Dim_Selected = dfDims[dfDims['Nr']==DID]
Dim_Selected
```

```python
# Szerokość rowka b => R
R = Dim_Selected['b_R'].iloc[0]
# R = dfDims['b_R'].iloc[DID-1]
print("Szerokość rowka (średnica freza): R = {0} [mm]".format(R))
```

```python
# Głębokość rowka t1 => G
G = Dim_Selected['t1_G'].iloc[0]
print("Głębokość rowka: G = {0} [mm]".format(G))
```

```python
# Losowanie średnicy wału (powierzchni cylindrycznej) w zadanym zakresie
Dmin = Dim_Selected['Dmin'].iloc[0]
Dmax = Dim_Selected['Dmax'].iloc[0]
print("Losowanie średnicy w zakresie od {0} do {1}".format(Dmin,Dmax))
D = np.random.randint(Dmin,Dmax)
print("Wylosowano średnicę D = {0} [mm]".format(D))
```

```python
# Losowanie długości wpustu w zadanym zakresie
Lmin = Dim_Selected['Lmin'].iloc[0]
Lmax = Dim_Selected['Lmax'].iloc[0]
print("Losowanie długości wpustu w zakresie od {0} do {1}".format(Lmin,Lmax))
```

```python
# Znormalizowane długości są w innej tabeli
dfLen.head()
```

```python
# Dyskretne wartości długości odczytujemy z innej tabeli wykonując selekcję dla danej szerokości rowka b -> R
Selected_Lens = dfLen[dfLen['b_R']==R]
Selected_Lens
```

```python
NRmin = Selected_Lens['Nr'].min()
NRmax = Selected_Lens['Nr'].max()
print("Losowanie w zakresie numerów od {0} do {1}".format(NRmin,NRmax))
NrID = np.random.randint(NRmin,NRmax)
print(NrID)
# Odczytaj długość L
NR_Selected = Selected_Lens[Selected_Lens['Nr']==NrID]
print(NR_Selected)
L = NR_Selected['L'].iloc[0]
print("Wylosowano długość L = {0} [mm]".format(L))
```

```python
print("Podsumowanie losowania")
print("Wymiary geometryczne rowka:")
print("Średnica wału D={0}, Szerokość rowka R={1}, Długość rowka L={2}, Głębokość rowka G={3}".format(D,R,L,G))
print("Parametry technologiczne:")
print("Pasowanie IT={0}, chropowatość Ra={1}".format(Fit,Ra))
print("Parametry materiałowe:")
print("Material PN: {0}, wytrzymałość Rm={1} [MPa], twardość HB={2}".format(Mat_PN, Rm, HB))
```
