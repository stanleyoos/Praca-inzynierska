```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Wygenerowane Dane.xlsx')
X = df[[ 'L', 'R', 'G', 'Rm', 'HB']]
y = df['pz']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
import pickle
with open('modelRF[pz]', 'wb') as f:
    pickle.dump(model,f)
```

```python
from sklearn.metrics import r2_score 
r2_score(y_test, y_predict)
```

```python
from sklearn.metrics import mean_squared_error, r2_score #
y_pred = model.predict(X_train)
lin_mse = mean_squared_error(y_train, y_pred)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(y_test, y_predict, alpha=0.6)

plt.xlabel('Wartość poprawna')
plt.ylabel('Wyliczona przez model')
plt.title('Losowy las fz [mm/ostrze] - 100 rowków')
```
