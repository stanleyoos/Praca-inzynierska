```python
import pandas as pd
```

```python
df = pd.read_excel('Wygenerowane Dane.xlsx')
X = df[[ 'L', 'R', 'G', 'Rm', 'HB']]
y = df['Vc']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train, y_train)
import pickle
with open('modelRF[Vc]', 'wb') as f:
    pickle.dump(model,f)
```

```python
y_predict = model.predict(X_test)
from sklearn.metrics import r2_score # współczynnik determinacji
r2_score(y_test, y_predict)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
plt.scatter(y_test, y_predict, alpha=0.1)

plt.xlabel('Wartość poprawna')
plt.ylabel('Wyliczona przez model')
plt.title('Regresja liniowa Vc [m/min] - 1000 rowków')
```

```python
pred_df = pd.DataFrame({'Actual value':y_test, 'Predicted value':y_predict, 'Difference': y_test-y_predict})
pred_df[0:20]
```
